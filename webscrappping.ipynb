````python
def web_scraping(urlpage,balise,attr,val,attr_output):
    #1-chargement de la page
    # req = Request(urlpage, headers={'User-Agent': 'Mozilla/5.0'})
    #webpage = urlopen(req, context=context).read()
    req = Request(urlpage, headers={'User-Agent': 'Mozilla/5.0'})
    page=urlopen(req).read()
    soup =BeautifulSoup(page,'html.parser')
    
    #2-chercher les link avec la class = colour colour+Mstart(8px)
    links = soup.find_all(balise, attrs={attr:val})
    output = []
    if attr_output != None:
        for i in range(len(links)):
            output.append(links[i][attr_output])
        page.close()
    else:
        output = links
    return output
````

````python
def get_links_page0(url):
 page=requests.get(url)
 soup =BeautifulSoup(page.content,'html.parser')
 liens = soup.find_all('a',{'class':"event-list",'href':True})
 link_final=[]
 link3_list=[]
 for lien in liens :
    link2= lien['href'].replace('\n', ' ').strip()
    link3_list.append(link2)
 for link in link3_list:
  link="https://www.eyebuydirect.fr/"+link
  link_final.append(link)
 return link_final
````

````python
def get_link_produit(url):
  page=requests.get(url)
  soup =BeautifulSoup(page.content,'html.parser')
  imgs = soup.find_all('img',{'data-bigsrc':True,'height':True})
  tag_list=[]
  link_list=[]
  for image in imgs :
    link2= image['data-bigsrc'].replace('\n', ' ').strip()
    height=image['height'].replace('\n', ' ').strip()
    if(image["height"]=="25"):
     tag_list.append(image)
  for i in tag_list:
   link_list.append(i['data-bigsrc'])
  return(link_list)
````

````python
def get_link_page(url):
  urls=get_links_page0(url)
  L=[]
  for i in urls:
    L.append(get_link_produit(i))
  return L
````

## Prix

````python
def get_prix_poduit(url):
  caract= web_scraping(urlpage=url,
             balise="span",attr="class",val="normal-price price-symbol",attr_output=None)
  caract = ' '.join(str(caract).split())
  start = str(caract).find('data-price=')
  end = str(caract)[start:].find(">")
  Marque = caract[start+len("data-price=") : start+end] 
  return Marque[1:5]
  
````
